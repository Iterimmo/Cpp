```python
# Считываем файл с заголовками
names = ['№', "X", "Y"]
for i in range(0,5000):
    names.append("v"+str(i))
names.extend(("K", "p1","p2","p3","p4"))

dataset_orig = pd.read_csv('signals.csv',names=names)
dataset = dataset_orig.copy()

# Стандартизируем наши X и Y (3896 = dataset['X'].mean()+dataset['Y'].mean())
dataset['X'] = dataset['X'] / 3896
dataset['Y'] = dataset['Y'] / 3896

# Разделяем размеченные и неразмеченные данные
dataset_train = dataset.loc[dataset.K== -1]
dataset_test = dataset.loc[dataset.K!= -1]


# Находим среднее всех отсчётов
dataset_mean = pd.DataFrame(dataset.drop(['X','Y','K','p1','p2','p3','p4'],axis=1).mean(axis=1),columns=['mean'])

# Создаем новый датафрейм на со столбцами X,Y и mean
dataset_true_mean = pd.concat([dataset[['X','Y']], dataset_mean],axis=1)

# Создаем отдельные датафреймы, содержащие в себе только те строки датасета, в которых характерные точки не равны -1
dataset_p1 = dataset_test[dataset_test.p1!=-1]
dataset_p2 = dataset_test[dataset_test.p2!=-1]
dataset_p3 = dataset_test[dataset_test.p3!=-1]
dataset_p4 = dataset_test[dataset_test.p4!=-1]

# Создаем отдельные датафреймы, содержащие в себе только те строки датасета, в которых характерные точки равны -1
dataset_p1_nones = dataset[dataset.p1==-1] 
dataset_p2_nones = dataset[dataset.p2==-1]
dataset_p3_nones = dataset[dataset.p3==-1]
dataset_p4_nones = dataset[dataset.p4==-1]

# Находим среднее в тех датасетах, в которых характерные точки не равны -1 и создаем на их основе новый датасет, включающий
# в себя столбцы X,Y и mean

dataset_mean_p1 = pd.DataFrame(dataset_p1.drop(['X','Y','K','p1','p2','p3','p4'],axis=1).mean(axis=1),columns=['mean'])
dataset_true_mean_p1 = pd.concat([dataset_p1[['X','Y']], dataset_mean_p1],axis=1)

dataset_mean_p2 = pd.DataFrame(dataset_p2.drop(['X','Y','K','p1','p2','p3','p4'],axis=1).mean(axis=1),columns=['mean'])
dataset_true_mean_p2 = pd.concat([dataset_p2[['X','Y']], dataset_mean_p2],axis=1)

dataset_mean_p3 = pd.DataFrame(dataset_p3.drop(['X','Y','K','p1','p2','p3','p4'],axis=1).mean(axis=1),columns=['mean'])
dataset_true_mean_p3 = pd.concat([dataset_p3[['X','Y']], dataset_mean_p3],axis=1)

dataset_mean_p4 = pd.DataFrame(dataset_p4.drop(['X','Y','K','p1','p2','p3','p4'],axis=1).mean(axis=1),columns=['mean'])
dataset_true_mean_p4 = pd.concat([dataset_p4[['X','Y']], dataset_mean_p4],axis=1)

# Находим среднее в тех датасетах, в которых характерные точки равны -1 и создаем на их основе новый датасет, включающий
# в себя столбцы X,Y и mean

dataset_p1_mean_nones = pd.DataFrame(dataset_p1_nones.drop(['X','Y','K','p1','p2','p3','p4'],axis=1).mean(axis=1),columns=['mean'])
dataset_true_mean_p1_nones = pd.concat([dataset_p1_nones[['X','Y']], dataset_p1_mean_nones],axis=1)

dataset_p2_mean_nones = pd.DataFrame(dataset_p2_nones.drop(['X','Y','K','p1','p2','p3','p4'],axis=1).mean(axis=1),columns=['mean'])
dataset_true_mean_p2_nones = pd.concat([dataset_p2_nones[['X','Y']], dataset_p2_mean_nones],axis=1)

dataset_p3_mean_nones = pd.DataFrame(dataset_p3_nones.drop(['X','Y','K','p1','p2','p3','p4'],axis=1).mean(axis=1),columns=['mean'])
dataset_true_mean_p3_nones = pd.concat([dataset_p3_nones[['X','Y']], dataset_p3_mean_nones],axis=1)

dataset_p4_mean_nones = pd.DataFrame(dataset_p4_nones.drop(['X','Y','K','p1','p2','p3','p4'],axis=1).mean(axis=1),columns=['mean'])
dataset_true_mean_p4_nones = pd.concat([dataset_p4_nones[['X','Y']], dataset_p4_mean_nones],axis=1)
```

```python
# Разделяем наши данные на X,Y и обучаем их на 4 регрессионных алгоритмах

X= pd.DataFrame(dataset_true_mean_p1[['X','Y','mean']])
y= pd.DataFrame(dataset_p1['p1'])


lr_p1 = LinearRegression()
lr_p1.fit(X, y)

rf_p1 = RandomForestRegressor(n_estimators=1000, max_depth=10)
rf_p1.fit(X, y.values.ravel())

dt_p1 = DecisionTreeClassifier()
dt_p1.fit(X, y)

lgb_p1 = LGBMRegressor()
lgb_p1.fit(X, y)

# Подсчитываем метрику
print(lr_p1.score(X, y))
print(rf_p1.score(X, y))
print(dt_p1.score(X, y))
print(lgb_p1.score(X, y))



# Единица склонна к переобучению. Взяли 0.92
p1_pred = rf_p1.predict(dataset_true_mean_p1_nones[['X','Y','mean']]).round(0)
p1_pred = p1_pred.astype(np.int64)
```

```python
# Формируем датасет
dataset_for_clas = pd.concat([dataset[['X','Y','p1','p2','p3','p4','K']], dataset_true_mean['mean']],axis=1)

# Разделяем весь наш датасет на датасет с размеченными данными и без размеченных данных
dataset_for_clas_test = dataset_for_clas[dataset_for_clas.K != -1]
dataset_for_clas_train = dataset_for_clas[dataset_for_clas.K == -1]

```

```python
# Выделяем X и Y и разделяем их ещё на X_train, X_test, y_train, y_test при помощи train_test_split

X = pd.DataFrame(dataset_for_clas_test[['X','Y','mean']])
y = pd.DataFrame(dataset_for_clas_test['K'])

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= .3, stratify=y, random_state=42)
```

```python
# Обучаем 4 модели для классификации и смотрим метрику. Берем алгоритм с самой большой метрикой - 0.848485

LR = LogisticRegression(
    random_state=42, solver='lbfgs', multi_class='multinomial',max_iter=8000).fit(X_train, y_train.values.ravel()) 
print(round(LR.score(X_test,y_test), 6))
 
SVM = svm.SVC(decision_function_shape="ovo").fit(X_train, y_train.values.ravel()) 
print(round(SVM.score(X_test, y_test), 6))
 
RF = RandomForestClassifier(n_estimators=1000, max_depth=10, random_state=42).fit(X_train, y_train.values.ravel()) 
print(round(RF.score(X_test, y_test), 6))
 
NN = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(150, 10), random_state=1).fit(
    X_train, y_train.values.ravel()) 
print(round(NN.score(X_test, y_test), 6))
```

```python
# Предсказываем значения

K_pred = RF.predict(dataset_for_clas_train[['X','Y','mean']])
```

```python
# Заполняем датасет предсказанынми значениями

dataset_copy_K = dataset.copy()


K_calc = 0
    
for i in range(len(dataset_copy_K)):
    
    if dataset_copy_K['K'][i] == -1:
        dataset_copy_K['K'][i] = K_pred[K_calc]
        K_calc += 1

dataset = dataset_copy_K
dataset
```
